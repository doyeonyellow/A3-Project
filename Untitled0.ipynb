## [문제] A 회사의 주가가 56,000원이고 B 회사의 주가가 125,000원이라고 할 때, 어떤 사람이 A주식 100주와 B주식 35주를 가지고 있다면 그 사람이 가지고 있는 주식의 총액을 계산하는 코드를 변수 선언 및 사용을 이용하여 작성하시오.

```python
# A 회사 주식 정보
stock_a_price = 56000  # A 회사 주가
stock_a_shares = 100   # A 회사 주식 수

# B 회사 주식 정보
stock_b_price = 125000 # B 회사 주가
stock_b_shares = 35    # B 회사 주식 수

# 각 회사의 주식 총액 계산
total_value_a = stock_a_price * stock_a_shares
total_value_b = stock_b_price * stock_b_shares

# 전체 주식 총액 계산
grand_total_value = total_value_a + total_value_b

# 결과 출력
print(f"A 회사 주식 총액: {total_value_a}원")
print(f"B 회사 주식 총액: {total_value_b}원")
print(f"총 주식 자산: {grand_total_value}원")
```

```python
A = 56000
B = 125000
A = int(input("A 회사 주가:"))
B = int(input("B 회사 주가:"))
total_money = (A * 100) + (B * 35)
print('총 주식금액:' + str(total_money))
print(f'총 주식금액:{total_money:,}원')
```

## [문제] 화폐단위 계산
<pre>
사용자로부터 금액을 입력 받아 화폐의 단위별 개수를 출력하는 프로그램을 작성하시오.
조건) 화폐의 종류: 만원, 천원, 백원, 십원 (일원 단위는 절사)
[실행 예]
금액을 입력하세요: 45630
만원: 4개
천원: 5개
백원: 6개
십원: 3개
</pre>

```python
num = int(input("금액을 입력하세요:"))

# 만원 단위 계산
man_won = num // 10000
print("만원:", man_won)

# 천원 단위 계산 (만원을 제외한 나머지 금액에서)
chun_won = (num % 10000) // 1000
print("천원:", chun_won)

# 백원 단위 계산 (천원을 제외한 나머지 금액에서)
baek_won = (num % 1000) // 100
print("백원:", baek_won)

# 십원 단위 계산 (백원을 제외한 나머지 금액에서)
ship_won = (num % 100) // 10
print("십원:", ship_won)
```

```python
money = int(input("금액을 입력하세요"))
만원 = money // 10000
천원 = (money%10000) // 1000
백원 = (money%1000) // 100
십원 = (money%100) // 10

print("만원:", 만원, "개")
print("천원:", 천원, "개")
print("백원:", 백원, "개")
print("십원:", 십원, "개")
```
